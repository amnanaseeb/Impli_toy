```python
# Spliting the dataset
ds_split = ds["validation"].train_test_split(test_size=0.3, seed=42)
train_dataset = ds_split["train"]
test_dataset = ds_split["test"]

# Defining label mappings
label_mapping_implicature = {'yes': 1, 'no': 0}
label_mapping_incoherent = {'yes': 1, 'no': 0}

# Preprocess the dataset
def preprocess(examples):
    # Combining utterance and response
    combined_input = examples["utterance"] + " " + examples["response"]
    model_inputs = tokenizer(combined_input, padding="max_length", truncation=True, max_length=128)

    # Mapping labels and add them to model inputs
    model_inputs["labels"] = label_mapping_implicature[examples["implicature"].lower()]
    model_inputs["incoherent_labels"] = label_mapping_incoherent[examples["incoherent_implicature"].lower()]
    return model_inputs

# Applying preprocessing to datasets
train_dataset = train_dataset.map(preprocess, batched=False)
test_dataset = test_dataset.map(preprocess, batched=False)

# Set format for PyTorch
train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
test_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])

```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

# training arguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=1e-2,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=4,
    weight_decay=0.01,
    gradient_accumulation_steps=2,
    logging_dir='./logs',
    logging_steps=10,
    fp16=True,
    bf16=False,
    dataloader_num_workers= 5,
    save_strategy="epoch",
    load_best_model_at_end=True,
    lr_scheduler_type="cosine",
    warmup_steps=500,
    save_steps=1000,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
)

trainer.train()
```

```python
import re
from sklearn.metrics import classification_report
import pandas as pd

tokenizer = XLMRobertaTokenizer.from_pretrained("/content/trained_xlm_roberta")
model = XLMRobertaForSequenceClassification.from_pretrained("/content/trained_xlm_roberta")
model.eval()
model.to('cuda' if torch.cuda.is_available() else 'cpu')  # Use GPU if available

# Loading file
file_path = "/content/just conversation.xlsx"
data = pd.read_excel(file_path)

# Concatenate 'utterance' and 'response' columns to classify them together
data['combined_text'] = data['utterance'] + " " + data['response']

# Preprocess and make predictions
def preprocess_and_predict(text):
    # Tokenize the text
    inputs = tokenizer(
        text,
        padding="max_length",
        truncation=True,
        max_length=128,
        return_tensors="pt"
    ).to('cuda' if torch.cuda.is_available() else 'cpu')

# Initializing an empty list to store results
results = []

# Classifing each combined text (utterance + response)
for combined_text in data['combined_text']:
    inputs = tokenizer(combined_text, return_tensors="pt", padding=True, truncation=True, max_length=512).to("cuda")
    outputs = model(**inputs)
    predicted_class = torch.argmax(outputs.logits, dim=1).item()

    #Appending the predicted class for each row
    results.append(predicted_class)
data['Predicted Implicature Class'] = data['combined_text'].apply(preprocess_and_predict)
data['Predicted Implicature Class'] = results
if 'implicature' in data.columns:
    true_labels = data['implicature'].map({'yes': 1, 'no': 0}).tolist()
    print(classification_report(true_labels, results))
data.to_csv('/content/predicted_implicature_results.csv', index=False)
data.head()
```
